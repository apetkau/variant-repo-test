```python
from pyroaring import BitMap
import pandas as pd

def size_of_bitmap(start: int, length: int) -> int:
    bm = BitMap()
    for i in range(start, start + length):
        bm.add(i)
    return len(bm.serialize())

cases_contiguous_df = pd.DataFrame([
    {
        'Start': 0,
        'Length': 10
    },
    {
        'Start': 0,
        'Length': 100
    },
    {
        'Start': 0,
        'Length': 1000
    },
    {
        'Start': 1000,
        'Length': 10
    },
    {
        'Start': 1000,
        'Length': 100
    },
    {
        'Start': 1000,
        'Length': 1000
    },
    {
        'Start': 1000000,
        'Length': 10
    },
    {
        'Start': 1000000,
        'Length': 100
    },
    {
        'Start': 1000000,
        'Length': 1000
    }
])

cases_contiguous_df['Bitmap size'] = cases_contiguous_df.apply(lambda x: size_of_bitmap(x['Start'], x['Length']), axis='columns')
cases_contiguous_df
```

```python
import random

cases_random_df = pd.DataFrame([
    {
        'Max': 1,
        'Length': 10
    },
    {
        'Max': 1,
        'Length': 100
    },
    {
        'Max': 1,
        'Length': 1000
    },
    {
        'Max': 1000,
        'Length': 10
    },
    {
        'Max': 1000,
        'Length': 100
    },
    {
        'Max': 1000,
        'Length': 1000
    },
    {
        'Max': 1000000,
        'Length': 10
    },
    {
        'Max': 1000000,
        'Length': 100
    },
    {
        'Max': 1000000,
        'Length': 1000
    },
    {
        'Max': 1000000000,
        'Length': 10000
    },
    {
        'Max': 1000000000,
        'Length': 100000
    },
    {
        'Max': 1000000000,
        'Length': 1000000
    }
])

def size_of_bitmap_random(max: int, length: int) -> int:
    bm = BitMap()
    for i in range(0, length):
        bm.add(random.randrange(max))
    return len(bm.serialize())

cases_random_df['Bitmap size'] = cases_random_df.apply(lambda x: size_of_bitmap_random(x['Max'], x['Length']), axis='columns')
cases_random_df
```
